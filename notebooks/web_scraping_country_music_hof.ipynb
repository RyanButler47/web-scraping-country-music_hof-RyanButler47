# Looping over all 16 website pages

```python
#For loop for multiple tables and append them together
data = []
for page in range(2,17):
    website_url = 'https://countrymusichalloffame.org/hall-of-fame/members/page/'  + str(page)
    response = requests.get(website_url)
    print(response.status_code)
    data.append(response.content)
print(len(data))

#response.status_code
```

```python
print(type(response))
data
```

```python
soup = BeautifulSoup(data, 'lxml')
print(soup.title)
```
